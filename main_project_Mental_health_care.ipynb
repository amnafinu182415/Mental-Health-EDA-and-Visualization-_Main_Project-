# **Data Cleaning**

## *üßπ  Data Pre-processing

```python
# Standardise Column Names
data.columns= data.columns.str.strip().str.replace(' ', '_')
data
```

### Convert  to Date Format

```python
data['Time_Period_Start_Date'] = pd.to_datetime(data['Time_Period_Start_Date'])
data['Time_Period_Start_Date'] 
```

```python
data['Time_Period_End_Date']=pd.to_datetime(data['Time_Period_End_Date'])
data['Time_Period_End_Date']
```

```python
# After Conversion
data[['Time_Period_Start_Date','Time_Period_End_Date']].dtypes
```

```python
print("-----------------------------")
print('Before Handling Missing Values')
print("-----------------------------")
print(data.isnull().sum()) 
```

```python
suppressed_count=data['Suppression_Flag'].notna().sum()
not_suppressed_count=data['Suppression_Flag'].isna().sum()

print("Suppressed rows:", suppressed_count)
print("Not suppressed rows:", not_suppressed_count)
```

```python
print("Total rows:", data.shape[0])
```

```python
#removes rows where data is suppressed and keeps only the good rows.
data = data[data['Suppression_Flag'].isna()]
data
```

```python
print("Total rows:", data.shape[0])
```

```python
print(data.isnull().sum())
```

```python
#  Drop the Column
data= data.drop('Suppression_Flag',axis=1)
data
```

```python
# check duplicates
dupes = data.duplicated().sum()
print(dupes)
```

```python
numerical_colms = data.select_dtypes(include=['float64', 'int64']).columns
numerical_colms
```

```python
categorical_colms = data.select_dtypes(include=['object']).columns
categorical_colms
```

```python
data[numerical_colms] =data[numerical_colms].fillna(data[numerical_colms].mean())
data[numerical_colms]
```

```python
data[categorical_colms] =data[categorical_colms].fillna(data[categorical_colms].mode().iloc[0])
data[categorical_colms]
```

```python
print("-----------------------------")
print('After Handling Missing Values')
print("-----------------------------")
print(data.isnull().sum()) 
```

## Derived New Columns

```python
# Add column
data['confidence_range'] = data['HighCI'] - data['LowCI']
data['confidence_range']
data
```

```python
# Extract Year from Date
data['year'] = (data['Time_Period_Start_Date']).dt.year
data
```

```python
data['value_category'] = pd.qcut(data['Value'], 4, labels=['Low','Medium-Low','Medium-High','High'])
data['value_category']
data
```

## Exploratory Data Analysis (EDA)

```python
# Aggregation
#1. Mean Value by State                                      # Average mental health care usage across each state.
mean_value_by_st=data.groupby('State')['Value'].mean()
mean_value_by_st
```

```python
# 2. Mean Value by Group
Mean_Value_byGrp =data.groupby('Group')['Value'].mean()
Mean_Value_byGrp                                           # Which demographic groups (Age, Gender, Race, etc.) receive more care.
```

```python
# Create a Month Column
data['Month'] = data['Time_Period_Start_Date'].dt.to_period('M').astype(str)
# Shows how mental-health care usage changes across states AND across months simultaneously.
pivot_state_month = data.pivot_table(values='Value',index='State',columns='Month',aggfunc='mean')
pivot_state_month
```

```python
#3. Highest State for Each Group
data.loc[data.groupby('Group')['Value'].idxmax()][['Group','State','Subgroup','Value']]
```

```python
#  Correlation Across Numeric Features
plt.figure(figsize=(10,6))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap - Numeric Features")
plt.show()
```

```python
# 4. Highest Value in Each State
data.loc[data.groupby('State')['Value'].idxmax()]#[['State','Group','Subgroup','Value']]
```

```python
# Shows how mental-health care usage changes across states AND across months simultaneously.
pivot_state_month = data.pivot_table(values='Value',index='State',columns='Month',aggfunc='mean')
pivot_state_month
```

```python
# Distribution of Confidence Range
plt.figure(figsize=(8,4))
sns.kdeplot(data['confidence_range'], fill=True)  #This fills the area under the KDE curve with color.
plt.title("Distribution of Confidence Range",fontweight='bold',fontsize=15,color='y')
plt.xlabel("Confidence Range")
plt.show()
```

```python
# Line Plot: Confidence Range Trend Over Time
CR_Time = data.groupby('Month')['confidence_range'].mean()

plt.figure(figsize=(14,4))
sns.lineplot(x=CR_Time.index, y=CR_Time.values, marker='*',ms=20,mec='r',mfc='y',linewidth = 5, color='k')
plt.xticks(rotation=90)
plt.title("Monthly Trend of Confidence Range",fontweight='bold',color='m')
plt.ylabel("Mean Confidence Range")
plt.show()
```

```python
#  Pair Plot - Value, LowCI, HighCI,confidence_range
sns.pairplot(data[['Value', 'LowCI', 'HighCI','confidence_range']],diag_kind="kde")
plt.suptitle("Pairwise Relationships", y=1.02,fontsize=15, color='purple') 
plt.show()
```

üåü INSIGHTS
* Pair Plot - Value, LowCI, HighCI,Confidence_Range
* Variables-->Value, LowCI, HighCI,confidence_range
* Diagonal (KDE plots): Show the distribution of each variable, helping to understand the range and density of values. Off-diagonal (Scatter plots):      Show pairwise relationships between variables.
* Value vs LowCI shows a positive trend, meaning higher mental-health usage comes with higher lower confidence bounds.
* The pairplot reveals strong internal consistency between Value, LowCI, and HighCI, indicating that these measurements increase together and follow      similar distributions.
* The confidence_range variable is weakly related to Value, showing that uncertainty does not depend directly on the mental-health care percentage.
